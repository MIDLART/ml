6. Как вычислить ((A+B)*(-A/2)) на месте (без копирования)?

````python
A = np.array([1, 2, 3, 4, 5], dtype=np.float64)
B = np.array([6, 7, 8, 9, 10], dtype=np.float64)

np.add(A, B, out=B)
np.negative(A, out=A)
np.divide(A, 2, out=A)
np.multiply(B, A, out=A)

print(A)
````

```
State - название штата/региона

```

````python
import pandas as pd
````

````python
df = pd.read_csv("CrimesOnWomenData.csv")
````

1. За какие года представлен набор данных?

````python
unique_years = df["Year"].unique()

print(f"Данные представлены за следующие года: {unique_years}")
````

2. Выявите район с самым высоким количеством убийств женщин

````python
state_wise_dd = df.groupby("State")["DD"].sum()

state_with_max_dd = state_wise_dd.idxmax()
max_dd = state_wise_dd.max()

print(state_with_max_dd)
````

3. Какое среднее число изнасилований женщин в индии наблюдалось в указанные года?

````python
yearly_rape_sums = df.groupby('Year')['Rape'].sum()

total_rapes = yearly_rape_sums.sum()
num_years = len(yearly_rape_sums)

average_rapes_per_year = total_rapes / num_years

print(average_rapes_per_year)
````

4. Выявите район с минимальным числом похищений за период 2001-2003 года

````python
df_filtered = df[(df['Year'] >= 2001) & (df['Year'] <= 2003)]

min_kidnapping = df_filtered.loc[df_filtered['K&A'].idxmin()]

print(min_kidnapping['State'])
````

5. Перечислите районы в которых наблюдались наиболее частые случае домашнего насилия за период с 2005 по 2012 года

````python
df_filtered = df[(df['Year'] >= 2005) & (df['Year'] <= 2012)]

domestic_violence_by_state = df_filtered.groupby('State')['DV'].sum()
top_5_states = domestic_violence_by_state.nlargest(5)

print(top_5_states)
````

6. Выведите названия районов в порядке не возрастания случаев торговли женщинами за 2020 год

````python
df_2020 = df[df['Year'] == 2020]

sorted_states = df_2020.sort_values('WT', ascending=False)

print(sorted_states['State'])
````

8. Какой средний годовой доход у клиентов с уровнем лояльности больше 5 живущих на севере

````python
filtered_df2 = df2[(df2['loyalty_score'] > 5) & (df2['region'] == 'North')]

average_income = filtered_df2['annual_income'].mean()

if pd.isna(average_income):
    print("Нет клиентов, удовлетворяющих условиям.")
else :
    print(f"Средний годовой доход: {average_income}")
````

9. Выведите 25 и 75 перцентили уровня лоальности клиентов живущих на западе с частотой покупок более 20 раз в год

````python
filtered_df2 = df2[(df2['region'] == 'West') & (df2['purchase_frequency'] > 20)]

percentile_25 = filtered_df2['loyalty_score'].quantile(0.25)
percentile_75 = filtered_df2['loyalty_score'].quantile(0.75)

print(f"25-й перцентиль: {percentile_25:.2f}")
print(f"75-й перцентиль: {percentile_75:.2f}")
````

10. Отличается ли медианный возраст клиентов с уровнем лояльности более 7 и менее 4 для жителей Востока и Севера.

````python
east_high_loyalty = df2[(df2['region'] == 'East') & (df2['loyalty_score'] > 7)]['age'].median()
east_low_loyalty = df2[(df2['region'] == 'East') & (df2['loyalty_score'] < 4)]['age'].median()

north_high_loyalty = df2[(df2['region'] == 'North') & (df2['loyalty_score'] > 7)]['age'].median()
north_low_loyalty = df2[(df2['region'] == 'North') & (df2['loyalty_score'] < 4)]['age'].median()

print(f"Медианный возраст клиентов с высоким уровнем лояльности (Восток): {east_high_loyalty}")
print(f"Медианный возраст клиентов с низким уровнем лояльности (Восток): {east_low_loyalty}")
print(f"Медианный возраст клиентов с высоким уровнем лояльности (Север): {north_high_loyalty}")
print(f"Медианный возраст клиентов с низким уровнем лояльности (Север): {north_low_loyalty}")

if (not pd.isna(east_high_loyalty)) and (not pd.isna(north_high_loyalty)):
    if (east_high_loyalty != north_high_loyalty):
        print("Медианный возраст клиентов с уровнем лояльности более 7 отличается")
    else:
        print("Медианный возраст клиентов с уровнем лояльности более 7 не отличается")
else:
    print("Нет данных по уровню лояльности более 7")

if (not pd.isna(east_low_loyalty)) and (not pd.isna(north_low_loyalty)):
    if (east_low_loyalty != north_low_loyalty):
        print("Медианный возраст клиентов с уровнем лояльности менее 4 отличается")
    else:
        print("Медианный возраст клиентов с уровнем лояльности менее 4 не отличается")
else:
    print("Нет данных по уровню лояльности менее 4")
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Постройте диаграмму которая отражает средний уровень лояльности во всех регионах

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('CustomerPurchasingBehaviors.csv')

average_loyalty_by_region = df.groupby('region')['loyalty_score'].mean()

plt.figure(figsize=(8, 6))
plt.bar(average_loyalty_by_region.index, average_loyalty_by_region.values)
plt.xlabel("Регион")
plt.ylabel("Средний уровень лояльности")
plt.title("Средний уровень лояльности в каждом регионе")
plt.show()
````

2. Постройте гистограмму распределения зарплат клиентов на Севере.

````python
north_customers = df[df['region'] == 'North']

plt.figure(figsize=(10, 6))
plt.hist(north_customers['annual_income'], bins=10, edgecolor='black')
plt.xlabel("Годовой доход")
plt.ylabel("Количество клиентов")
plt.title("Распределение годового дохода клиентов на Севере")
plt.show()
````

3. Сравните уровни зарплат людей на севере и юге.

````python
north_customers = df[df['region'] == 'North']['annual_income']
south_customers = df[df['region'] == 'South']['annual_income']

plt.figure(figsize=(10, 6))
plt.hist(north_customers, bins=10, edgecolor='black', alpha=0.5, label='Север')
plt.hist(south_customers, bins=10, edgecolor='black', alpha=0.5, label='Юг')
plt.xlabel("Годовой доход")
plt.ylabel("Количество клиентов")
plt.title("Сравнение распределения годового дохода на Севере и Юге")
plt.legend(loc='upper right')
plt.show()
````

4. Постройте box plot отображающий распределение уровня лояльности для людей до 20 лет, от 20 до 40, от 40 до 60 и старше 60 лет

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bins = [0, 20, 40, 60, float('inf')]
labels = ['До 20', '20-40', '40-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='loyalty_score', data=df)
plt.xlabel("Возрастная группа")
plt.ylabel("Уровень лояльности")
plt.title("Распределение уровня лояльности по возрастным группам")
plt.show()
````
